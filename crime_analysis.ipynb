```python
import matplotlib.pyplot as plt
import pandas as pd

crimes = pd.read_csv('crimes_india_2001_2012.csv')
```

```python
crimes.head()
```

```python
crimes_in_2001 = crimes[crimes['year'] == 2001]
crimes_in_2001.head()
```

```python
crimes_in_2001.shape
```

```python
crimes.columns
```

```python
col = ['murder', 'robbery', 'riots', 'cheating',
       'dowry_death', 'cruelty_by_hus_rel', 'importation_of_girls', 'rape',
       'kidnapping', 'theft', 'total']
```

```python
crimes_by_year = crimes.groupby('year')[col].sum()
```

```python
crimes_by_year.head(12)
```

```python
crimes_by_year.reset_index(inplace=True)
crimes_by_year
```

```python
plt.plot(crimes_by_year['year'], crimes_by_year['total'])
plt.show()
```

```python
plt.plot(crimes_by_year['year'] , crimes_by_year['total'])
plt.xlabel('Years')
plt.ylabel('Total no of crimes')
plt.title('Crimes In India Trend, 2001-2012')
plt.show()
```

```python
plt.savefig('graph.png')
```

```python
plt.plot(crimes_by_year['year'], crimes_by_year['murder'], color='red' , label='MURDER')
plt.plot(crimes_by_year['year'], crimes_by_year['rape'] , color='blue', label='RAPE')
plt.xlabel('Years')
plt.ylabel('Murder and Rape crimes')
plt.title('Murder and Rape crime Trend , 2001-2012')
plt.legend()
plt.show()
```

```python
plt.savefig('p.png')
```

```python
fig = plt.figure(figsize =(10,6))
ax1 = fig.add_subplot(2,2,1)
ax1.plot(crimes_by_year['year'], crimes_by_year['murder'],color='red')
ax1.set_title('MURDER')

ax2 = fig.add_subplot(2,2,2)
ax2.plot(crimes_by_year['year'], crimes_by_year['rape'],color='pink')
ax2.set_title('RAPE')

ax3 = fig.add_subplot(2,2,3)
ax3.plot(crimes_by_year['year'], crimes_by_year['robbery'],color='black')
ax3.set_title('ROBBERY')

ax4 = fig.add_subplot(2,2,4)
ax4.plot(crimes_by_year['year'], crimes_by_year['dowry_death'])
ax4.set_title('DOWERY DEATH')

plt.tight_layout()
plt.show()
```

```python
crimes.columns
```

```python
cols =['murder', 'robbery', 'riots', 'cheating',
       'dowry_death', 'cruelty_by_hus_rel', 'importation_of_girls', 'rape',
       'kidnapping', 'theft', 'total']

crimes_by_state = crimes.groupby('state')[cols].sum()
crimes_by_state.head(10)
```

```python
crimes_by_state.sort_values(by=['total'], ascending=False, inplace=True)
crimes_by_state.head()
```

```python
# bar l=plot
from numpy import arange
x_pos = arange(10)

# Height of the bars
bar_height = crimes_by_state.iloc[0:10, 10]

# create the bar plot
plt.bar(x_pos, bar_height, .5,)

# positions and labels of x-ticks
plt.xticks(x_pos, crimes_by_state.index[0:10], rotation=90)
plt.xlabel('State')
plt.ylabel('Total Number of crimes')
plt.title('Top 10 States in India For Crime, 2001-2012')
plt.show()
```

```python
# Horizontal Bar plot
fig = plt.figure(figsize=(8,5))
y_pos = arange(5)
bar_width = crimes_by_state.iloc[0:5,10]
plt.barh(y_pos, bar_width, 0.5, color='green')
plt.yticks(y_pos, crimes_by_state.index[0:5])

plt.xlabel('Total number of crime')
plt.ylabel('States')
plt.title('Top 10 States in India For Crime, 2001-2012')
plt.show()
```

```python
# grouped bar plot

# Make the size of the plot bigger
fix = plt.figure(figsize=(10,6))

# positions of bars
x_pos = arange(5) 

# width of the bars
width = 0.20

# plot the bar for Murder
plt.bar(x_pos, crimes_by_state.iloc[0:5, 0], width,color='blue', label='Murder')

#plot the bar for cheating
plt.bar(x_pos + width , crimes_by_state.iloc[0:5,3] , width , color='orange',label='Cheating')

 


```
